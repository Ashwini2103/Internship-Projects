```python
# Importing Libraries
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize,RegexpTokenizer
from spellchecker import SpellChecker
import string
```

```python
# Path of the text file
data=pd.read_excel(r'C:\Users\Dad\Documents\Language SpellChecker\English-testcases.xlsx')
```

```python
# Dataframe content
data
```

```python
# Finding out the misspelled words in the sentence and also loading a text file as a word_dictionary apart from inbuild frequency list
misspelled_words=[]
from collections import Counter
for text in data['English -  spellings and grammar issues']:
    text=text.lower()
    tokenizer = RegexpTokenizer("[\w']+")
    words=tokenizer.tokenize(text)
    spell=SpellChecker()
    spell.word_frequency.load_text_file(r'C:\Users\Dad\Documents\Language SpellChecker\my_free_text_doc.txt')
    misspelled_words.append( spell.unknown(words) )
    
```

```python
misspelled_words
```

```python
# Correction of words which are misspelled and storing the values in a dictionary
corrected_words=[]
dict_keys={}
for idx,misspell in enumerate(misspelled_words):
    misspell = list(misspell)
    if misspell == []:
        misspell.append('-')
    for index,text in enumerate(misspell):
        dict_keys[text]=spell.candidates(text)
        
        
```
