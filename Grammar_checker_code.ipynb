```python
from gingerit.gingerit import GingerIt
import pandas as pd
```

```python
# Path of the text file
data=pd.read_excel(r'C:\Users\Dad\Documents\Language SpellChecker\English-testcases.xlsx')
```

```python
parser = GingerIt()
```

```python
corrected_sentences = []
for i in data['English -  spellings and grammar issues']:
    try:
        string_length=len(i)+5  # will be adding 5 extra spaces
        string_revised=i.ljust(string_length)
        corrected_sentences.append(parser.parse(string_revised))
    except:
        print(" ")
        
```
